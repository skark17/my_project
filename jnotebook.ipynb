```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

data_frame = pd.read_csv('static/mushrooms.csv')
data_frame = data_frame.drop(["veil-type"],axis=1)

labels = data_frame['class'].value_counts()
# Defines the mapping of categories to ordinal values.
lc=LabelEncoder()
for column in data_frame.columns:
    data_frame[column] = lc.fit_transform(data_frame[column])    
# Creates a pie chart that determines toxic and non-toxic mushrooms.
plt.figure(figsize=(6, 6))
plt.pie(labels, labels=labels.index, autopct='%1.1f%%', startangle=140)
# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')  
plt.title('Toxic and Non-toxic Mushrooms')
plt.show()

# Creates a heat map where individual values in a matrix are represented as colors.
plt.figure(figsize=(16, 12))
sns.heatmap(data_frame.corr(), annot=True, annot_kws={"size": 8}, linewidths=.8,cmap="coolwarm")
selectDataFrame = pd.melt(
    data_frame, 'class', var_name='Characteristics')
X = data_frame.drop(['class'], axis=1)
Y = data_frame["class"]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=50, test_size=0.8)
plt.show()

# Machine learning method to learn and make predictions or decisions.
random_forest_classifier = RandomForestClassifier()
random_forest_classifier.fit(X_train, Y_train)
# Confusion matrix assessing the performance of a classification model.
forecast = random_forest_classifier.predict(X_test) 
confusionMatrix = confusion_matrix(Y_test, forecast)
horizontal_axis = ["Non-toxic", "Toxic"]
y_axis_labels = ["Non-toxic", "Toxic"]
f, axis = plt.subplots(figsize =(7,5))
cmap = "YlOrRd"
sns.heatmap(confusionMatrix, annot = True, linewidths=1, linecolor="white", fmt = ".0f",cmap=cmap, xticklabels=horizontal_axis, yticklabels=y_axis_labels)
plt.xlabel("Forecast")
plt.ylabel("Actual")
plt.title('Confusion Matrix');
plt.show()

predictions = random_forest_classifier.predict(X_test)
cr = classification_report(Y_test, predictions)
print(cr)

# Bar graph count of non-toxic/toxic.
count = data_frame['class'].value_counts()
plt.figure(figsize=(3, 2))
sns.barplot(x=count.index, y=count.values, palette="deep")
plt.ylabel('Count')
plt.xlabel('Class')
plt.title('Display of non toxic and toxic mushroom.')
plt.show() 

# Bar graph to visually represent categorical data.
features_list = X.columns.values
feature = random_forest_classifier.feature_importances_
sorted_idx = np.argsort(feature) 
plt.figure(figsize=(9,8))
plt.barh(range(len(sorted_idx)), feature[sorted_idx], color='blue')
plt.yticks(range(len(sorted_idx)), features_list[sorted_idx])
plt.xlabel('Classification Significance')
plt.title('Key Characteristics')
plt.draw()
plt.show()

# Violin plot to display the median, quartiles, and the overall distribution of the data.
selectDataFrame = pd.melt(data_frame, 'class', var_name='Characteristics')
figure, axis = plt.subplots(figsize=(10,7))
custom_palette = sns.color_palette(["skyblue", "salmon"])
a = sns.violinplot(axis = axis, x='Characteristics', y='value', hue='class', split = True, data=selectDataFrame, inner = 'quartile', palette = custom_palette)
df_no_class = data_frame.drop(['class'],axis = 1)
a.set_xticklabels(rotation = 90, labels = list(df_no_class.columns));


# Cap color horizontal bar plot
total = 5
myDataFrame = pd.read_csv("static/mushrooms.csv") # Initiating data frame 
characteristic = myDataFrame[['class', 'cap-color']]

# Red
EP = len(characteristic.loc[(characteristic['cap-color'] == 'e') & (characteristic['class'] == 'p')])
EE = len(characteristic.loc[(characteristic['cap-color'] == 'e') & (characteristic['class'] == 'e')])
# Gray
GP = len(characteristic.loc[(characteristic['cap-color'] == 'g') & (characteristic['class'] == 'p')])
GE = len(characteristic.loc[(characteristic['cap-color'] == 'g') & (characteristic['class'] == 'e')])

# Brown
NP = len(characteristic.loc[(characteristic['cap-color'] == 'n') & (characteristic['class'] == 'p')])
NE = len(characteristic.loc[(characteristic['cap-color'] == 'n') & (characteristic['class'] == 'e')])

# White
WP = len(characteristic.loc[(characteristic['cap-color'] == 'w') & (characteristic['class'] == 'p')])
WE = len(characteristic.loc[(characteristic['cap-color'] == 'w') & (characteristic['class'] == 'e')])
# Yellow
YP = len(characteristic.loc[(characteristic['cap-color'] == 'y') & (characteristic['class'] == 'p')])
YE = len(characteristic.loc[(characteristic['cap-color'] == 'y') & (characteristic['class'] == 'e')])
non_toxic = ( EE, GE, NE, WE, YE)
toxic = ( EP, GP, NP, WP, YP)
ind = np.arange(total)
width = 0.35
figure = plt.figure()

categories = ('Red', 'Gray','Brown', 'White', 'Yellow')
plt.figure(figsize=(10, 8))
plt.barh(categories, non_toxic, color='purple')
plt.barh(categories, toxic, left=non_toxic, color='blue')
plt.xlabel('Count')
plt.title('Class by Cap Color Attributes')
plt.legend(labels=['Non-toxic', 'Toxic'])
plt.show()

# Cap shape horizontal bar plot.
total = 4
myDataFrame = pd.read_csv("static/mushrooms.csv")
characteristic = myDataFrame[['class', 'cap-shape']]
# Bell
BP = len(characteristic.loc[(characteristic['cap-shape'] == 'b') & (characteristic['class'] == 'p')])
BE = len(characteristic.loc[(characteristic['cap-shape'] == 'b') & (characteristic['class'] == 'e')])

# Flat
FP = len(characteristic.loc[(characteristic['cap-shape'] == 'f') & (characteristic['class'] == 'p')])
FE = len(characteristic.loc[(characteristic['cap-shape'] == 'f') & (characteristic['class'] == 'e')])

# Sunken
SP = len(characteristic.loc[(characteristic['cap-shape'] == 's') & (characteristic['class'] == 'p')])
SE = len(characteristic.loc[(characteristic['cap-shape'] == 's') & (characteristic['class'] == 'e')])

# Convex
XP = len(characteristic.loc[(characteristic['cap-shape'] == 'x') & (characteristic['class'] == 'p')])
XE = len(characteristic.loc[(characteristic['cap-shape'] == 'x') & (characteristic['class'] == 'e')])

non_toxic = ( BE, FE, SE, XE)
toxic = ( BP, FP, SP, XP)
ind = np.arange(total) # the x locations for the groups
width = 0.35
figure = plt.figure()

categories = ('Bell', 'Flat','Sunken', 'Convex')
plt.figure(figsize=(10, 8))
plt.barh(categories, non_toxic, color='purple')
plt.barh(categories, toxic, left=non_toxic, color='blue')
plt.xlabel('Count')
plt.title('Class by Cap-shape Attributes')
plt.legend(labels=['Non-toxic', 'Toxic'])
plt.show()

# Cap surface horizontal BAR PLOT
total = 4
myDataFrame = pd.read_csv("static/mushrooms.csv")
characteristic = myDataFrame[['class', 'cap-surface']]
# fibrous
FP = len(characteristic.loc[(characteristic['cap-surface'] == 'f') & (characteristic['class'] == 'p')])
FE = len(characteristic.loc[(characteristic['cap-surface'] == 'f') & (characteristic['class'] == 'e')])

# Grooves
GP = len(characteristic.loc[(characteristic['cap-surface'] == 'g') & (characteristic['class'] == 'p')])
GE = len(characteristic.loc[(characteristic['cap-surface'] == 'g') & (characteristic['class'] == 'e')])
# Scaly
YP = len(characteristic.loc[(characteristic['cap-surface'] == 'y') & (characteristic['class'] == 'p')])
YE = len(characteristic.loc[(characteristic['cap-surface'] == 'y') & (characteristic['class'] == 'e')])

# Smooth
SP = len(characteristic.loc[(characteristic['cap-surface'] == 's') & (characteristic['class'] == 'p')])
SE = len(characteristic.loc[(characteristic['cap-surface'] == 's') & (characteristic['class'] == 'e')])

non_toxic = (FE, GE, YE, SE)
toxic = (FP, GP, YP, SP)
ind = np.arange(total) # the x locations for the groups
width = 0.35
figure = plt.figure()

categories = ('fibrous', 'grooves','scaly', 'smooth')
plt.figure(figsize=(10, 8))
plt.barh(categories, non_toxic, color='purple')
plt.barh(categories, toxic, left=non_toxic, color='blue')
plt.xlabel('Count')
plt.title('Class by Cap surface Attributes')
plt.legend(labels=['Non-toxic', 'Toxic'])
plt.show()

```
